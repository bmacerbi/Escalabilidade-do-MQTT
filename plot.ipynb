```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_data(file):
    # Load the log file
    log_file_path = file
    data = []

    # Read the log file
    with open(log_file_path, 'r') as file:
        for line in file:
            if 'CPU' in line and 'Memory' in line:
                parts = line.strip().split(', ')
                cpu_usage = float(parts[0].split(': ')[1].replace('%', ''))
                memory_usage = float(parts[1].split(': ')[1].replace(' Bytes', ''))
                data.append({'CPU': cpu_usage, 'Memory': memory_usage / (1024 ** 2)})

    # Convert data to DataFrame
    df = pd.DataFrame(data)

    # Plot the data
    plt.figure(figsize=(12, 6))

    # Plot CPU usage
    plt.subplot(2, 1, 1)
    plt.plot(df.index, df['CPU'], label='CPU Usage (%)', color='blue')
    plt.xlabel('Time (Index)')
    plt.ylabel('CPU Usage (%)')
    plt.title('CPU Usage Over Time')
    plt.legend()
    plt.grid(True)

    # Plot Memory usage
    plt.subplot(2, 1, 2)
    plt.plot(df.index, df['Memory'], label='Memory Usage (MB)', color='green')
    plt.xlabel('Time (Index)')
    plt.ylabel('Memory Usage (MB)')
    plt.title('Memory Usage Over Time')
    plt.legend()
    plt.grid(True)

    # Show the plots
    plt.tight_layout()
    plt.show()
```

```python
plot_data("broker_usage_publish.log")
```

```python
plot_data("broker_usage_ps.log")
```
